```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Load the dataset
file_path = "diabetes.csv"
df = pd.read_csv(file_path)

# Columns where zero is not a valid value
invalid_cols = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

# Replace 0s with NaN
df[invalid_cols] = df[invalid_cols].replace(0, np.nan)

# Check how many missing values now
print("ðŸ”¹ Missing values per column:")
print(df.isna().sum())

# Fill missing values with the median of each column
df[invalid_cols] = df[invalid_cols].fillna(df[invalid_cols].median())

# Confirm missing values are handled
print("\nðŸ”¹ Missing values after imputation:")
print(df.isna().sum())
```

```python
# Create Age Category
def age_category(age):
    if age < 31:
        return "Young"
    elif age <= 50:
        return "Middle-aged"
    else:
        return "Old"

df["AgeCategory"] = df["Age"].apply(age_category)

# Create BMI Category
def bmi_category(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif bmi < 25:
        return "Normal"
    elif bmi < 30:
        return "Overweight"
    else:
        return "Obese"

df["BMICategory"] = df["BMI"].apply(bmi_category)

# Preview new columns
print("ðŸ”¹ Sample of new features:")
display(df[["Age", "AgeCategory", "BMI", "BMICategory"]].head())
```
